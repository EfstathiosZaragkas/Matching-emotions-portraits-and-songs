```python
import pandas as pd
import shutil
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1, 1))

annotations_dir = "../PMEmo/PMEmo2019/annotations/static_annotations.csv"
songs_dir = "../PMEmo/PMEmo2019/chorus"
target_dir = "../Quadrant Songs Library"

q1_dir = "../Quadrant Songs Library/Q1/Modern"
q2_dir = "../Quadrant Songs Library/Q2/Modern"
q3_dir = "../Quadrant Songs Library/Q3/Modern"
q4_dir = "../Quadrant Songs Library/Q4/Modern"
n_dir = "../Quadrant Songs Library/N/Modern"
```

```python
df_annotations = pd.read_csv('../PMEmo/PMEmo2019/annotations/static_annotations.csv')
```

```python
df_annotations.head(5)
```

```python
df_annotations[['Arousal(mean)', 'Valence(mean)']] = scaler.fit_transform(df_annotations[['Arousal(mean)', 'Valence(mean)']])
```

```python
for index, row in df_annotations.iterrows():
    arousal = row['Arousal(mean)']
    valence = row['Valence(mean)']

    file_name = f"{int(row['musicId'])}"
    source_file_path = f"{songs_dir}/{file_name}.mp3"
    
    if(abs(arousal)< 0.25 and abs(valence)<0.25):
        pasted_file_name = f"{n_dir}/pmemo_{int(row['musicId'])}.mp3"
        shutil.copy(source_file_path, pasted_file_name)
        print(n_dir)
    elif(arousal>0 and valence>0):
        pasted_file_name = f"{q1_dir}/pmemo_{int(row['musicId'])}.mp3"
        shutil.copy(source_file_path, pasted_file_name)
        print(q1_dir)
    elif(arousal>0 and valence<0):
        pasted_file_name = f"{q2_dir}/pmemo_{int(row['musicId'])}.mp3"
        shutil.copy(source_file_path, pasted_file_name)
        print(q2_dir)
    elif(arousal<0 and valence<0):
        pasted_file_name = f"{q3_dir}/pmemo_{int(row['musicId'])}.mp3"
        shutil.copy(source_file_path, pasted_file_name)
        print(q3_dir)
    elif(arousal<0 and valence>0):
        pasted_file_name = f"{q4_dir}/pmemo_{int(row['musicId'])}.mp3"
        shutil.copy(source_file_path, pasted_file_name)
        print(q4_dir)
    
    
    
```
