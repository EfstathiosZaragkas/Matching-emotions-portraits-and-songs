```python
import pandas as pd
import shutil
import re 

arousal_annotations_dir = "../CCMED_WCMED/CCMED_WCMED/WCMED/WCMED-Ratings/WCMED_Arousal_Rating.csv"
valence_annotations_dir = "../CCMED_WCMED/CCMED_WCMED/WCMED/WCMED-Ratings/WCMED_Valence_Rating.csv"

songs_dir = "../CCMED_WCMED/CCMED_WCMED/WCMED/songs"
target_dir = "../Quadrant Songs Library"

q1_dir = "../Quadrant Songs Library/Q1/Classical"
q2_dir = "../Quadrant Songs Library/Q2/Classical"
q3_dir = "../Quadrant Songs Library/Q3/Classical"
q4_dir = "../Quadrant Songs Library/Q4/Classical"
n_dir = "../Quadrant Songs Library/N/Classical"
```

```python
df_arousal = pd.read_csv(arousal_annotations_dir)
df_valence = pd.read_csv(valence_annotations_dir)

df_arousal.rename(columns = {'rating':'arousal'}, inplace = True)
df_valence.rename(columns = {'rating':'valence'}, inplace = True)
```

```python
df_valence.groupby('name')[['valence']].mean()
```

```python
df_arousal.groupby('name')[['arousal']].mean()
```

```python
df_merged = pd.merge(df_arousal, df_valence, on='name')
df_merged.head()
```

```python
pattern = r'(_\d+)?\.mp3'

for index, row in df_merged.iterrows():
    arousal = row['arousal']
    valence = row['valence']

    section_name = f"{row['name']}"
    file_name = re.sub(pattern, '.mp3', section_name)
    source_file_path = f"{songs_dir}/{file_name}"
    
    pasted_file_name = f"{n_dir}/wcmed_{file_name}"
    
    if(abs(arousal)< 0.25 and abs(valence)<0.25):
        try:
            shutil.copy(source_file_path, pasted_file_name)
        except FileNotFoundError:
            continue
    elif(arousal>0 and valence>0):
        try:
            shutil.copy(source_file_path, pasted_file_name)
        except FileNotFoundError:
            continue
    elif(arousal>0 and valence<0):
        try:
            shutil.copy(source_file_path, pasted_file_name)
        except FileNotFoundError:
            continue
    elif(arousal<0 and valence<0):
        try:
            shutil.copy(source_file_path, pasted_file_name)
        except FileNotFoundError:
            continue
    elif(arousal<0 and valence>0):
        try:
            shutil.copy(source_file_path, pasted_file_name)
        except FileNotFoundError:
            continue
```
